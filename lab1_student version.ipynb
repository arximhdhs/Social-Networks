### Γ) Δίκτυα με βάρη

Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να δημιουργηθεί τυχαίος πίνακας βαρών. Το εύρος των βαρών να κυμαίνεται στο διάστημα $[1,10]$.

```python
# για τιμές στο [1,10]: α=1, b=10, weight=np.random.rand()*(b-a) + a
# για την προσθήκη βαρών στις ακμές, χρησιμοποήστε την add.edge του networkx

def add_random_weights_to_graph(H,a,b):
    G = H.copy();
    for j in G:
        for i in G[j]:
            G.add_edge(j,i,weight=np.random.rand()*(b-a) + a)
    return G

NREG = add_random_weights_to_graph(REG,1,10)
NRGER = add_random_weights_to_graph(RGER,1,10)
NRGG = add_random_weights_to_graph(RGG,1,10)
NSF = add_random_weights_to_graph(SF,1,10)
NSW = add_random_weights_to_graph(SW,1,10)
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α με τα βάρη στις ακμές:
#python 3 implementation

def strength(topology):
    weights_list=[]
    for i in range(0,len(topology.adj)):
        a=[]
        for j in range (0,len(topology.adj[i].items())):
            d=list(topology.adj[i].values())[j]
            c=list(d.values())[0]
            a.append(c)
        weights_list.append(a)
    node_weights = []
    for node in weights_list:
        node_weights.append(sum(node))
    return node_weights

REG_strength= strength(NREG)
RGER_strength=strength(NRGER)
RGG_strength=strength(NRGG)
SF_strength=strength(NSF)
SW_strength=strength(NSW)

sub = [REG_strength,RGER_strength,RGG_strength,SF_strength,SW_strength]
names=["REG","RGER","RGG","SF","SW"]
def plot_cumulative(lst,names):
    fig, ax = plt.subplots(nrows=1, ncols=5,figsize=(20,5))
    for i in range(len(lst)):
        temp = lst[i].copy()
        temp.sort()
        ax[i].plot(temp,[i for i in range(1,len(temp)+1)])
        ax[i].set_title(names[i])

```

```python
def compute_cc(G):
    total=[]
    for m in range(len(G)):
        sum1,d=0,{}
        x=list(G.adj[m])
        k=[list(G.adj[i]) for i in x]
        for i in x:
            d[i] = True
        for i in k:
            for l in i:
                if l in d:
                    sum1+=1
        if len(k)>1:
            final = sum1/(len(k)*(len(k)-1))
        else:
            final = float("Nan")
        total.append(final)
    return total
```

### Κ) Εγω-κεντρικότητες (Ego-centralities)
Για τις συνθετικές τοπολογίες του πίνακα 1, να υπολογιστεί η ενδιαμεσική εγω-κεντρικότητα (ego-betweenness centrality) του κάθε κόμβου. Στη συνέχεια να καταταχθούν οι κόμβοι με βάση την ενδιαμεσική εγωκεντρικότητα και την ενδιαμεσική κεντρικότητα και να συγκριθούν/σχολιαστούν οι κατατάξεις ως προς την ομοιότητά τους για κάθε μία τοπολογία.  

```python
#για την κατασκευή των εγωδικτύων, χρησιμοποιήστε την ego_graph του networkx
#για τον υπολογισμό της ενδιαμεσικής εγωκεντρικότητας θα χρειαστεί να δημιουργήσετε τον πίνακα γειτνίασης του εγωδικτύου
#Χρήσιμες συναρτήσεις: neighbors(networkx), ones(numpy),linalg.matrix_power, multiply, triu, nonzero του numpy
#κατασκευάστε dataframe από το pandas με γραμμές τους κόμβους και στήλες τις τιμές των υπό εξέταση κεντρικοτήτων
#κατασκευάστε στη συνέχεια pairplot από το seaborn για να εξετάσετε τις συσχετίσεις μεταξύ τους.
```

```python
def ego_calculate(G):
    ret = []
    for i in range(len(G)):
        G1 = nx.ego_graph(G,i)
        AG1 = nx.to_numpy_matrix(G1)
        nb  = AG1.shape[1]
        con = (nb-1)*(nb-2)/2
        sq = np.inner(AG1,AG1)
        shape1 = sq.shape
        ones = np.ones(shape1)
        o_m_A= ones - AG1
        final = np.triu(np.multiply(sq,o_m_A),+1)
        for x in range(len(final)):
            for k in range(len(final[x])):
                if final[x,k]!=0:
                    final[x,k] = 1/final[x,k]
        ret.append(final[1:,:].sum())
    return ret


sub = [
        [ego_calculate(REG),REG_betweenness_centrality],
          [ego_calculate(RGG),RGG_betweenness_centrality],
      [ego_calculate(RGER),RGER_betweenness_centrality],
        [ego_calculate(SF),SF_betweenness_centrality],
        [ego_calculate(SW),SW_betweenness_centrality]
      ]


# fig, ax = plt.subplots(nrows=1, ncols=5,figsize=(300,20))
# for ind,a in enumerate(ax):
#     a.plot(sub[ind][0],sub[ind][1],'o',color='blue')
#     a.set_ylabel("Number")
#     a.set_xlabel("Degree")
#     a.set_title(sub[ind][1])
# plt.show()
# k = ego_calculate(RGG)
# l = RGG_betweenness_centrality
# plt.plot(k,l,'o', color='black')
```
